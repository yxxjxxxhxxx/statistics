```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine

# 한글 폰트 사용을 위해서 세팅
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/H2GPRM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

wine_load = load_wine() # load_wine 함수를 변수에 넣기
# wine_load 하면 안에거 쭈우우욱
```

```python
wine_load.data #  wine_load 안에 있는 정보중에 data 만 출력
```

```python
wine_load.target # wine_load 안에 있는 target 만 출력
```

```python
wine = pd.DataFrame(wine_load.data, 
                    columns = wine_load.feature_names)
# 그냥 두면 보기 어려우니까 pd.DataFrame 사용
# wind_load.data 를 보기쉽게 만들고
# 위에다 columns 를 안쓰면 숫자가 나오는데 
# wind_load.feature_names

# 'feature_names': ['alcohol',
#   'malic_acid',
#   'ash',
#   'alcalinity_of_ash',
#   'magnesium',
#   'total_phenols',
#   'flavanoids',
#   'nonflavanoid_phenols',
#   'proanthocyanins',
#   'color_intensity',
#   'hue',
#   'od280/od315_of_diluted_wines',
#   'proline']}
```

```python
wine
```

```python
wine['Class'] = wine_load.target
```

```python
wine # 컬럼에 추가된 Class 를 확인하자
```

```python
# class 0, 1, 2 를 보기 좋게 바꿔봐라
wine = wine.replace({'Class':0}, 'A')
wine = wine.replace({'Class':1}, 'B')
wine = wine.replace({'Class':2}, 'C')
wine
```

```python
# wine['Class'] = wine['Class'].map({'class_0':'A',
#                                    'class_1':'B',
#                                    'class_2':'C'})
# map 을 써서 바꿀 수 도 있따.
```

```python
wine['Class'].value_counts() 
# 와인 정보에 있는 Class 중 값의 갯수 
```

```python
# 스캐터, 파이, 스캐터+플랏
```

```python
wine_type = wine['Class'].value_counts()
wine_type
```

```python
# list(wine_type.values)
# wine_type.index.tolist()
list(wine_type.values)
```

```python
plt.bar(x = wine_type.index, height = wine_type.values,
       width = 0.8,
       bottom = None, 
       align = 'edge', color=['r', 'g', 'b'])
plt.show()
```

```python
plt.barh(wine_type.index, wine_type.values,
         height = 0.8,
         left = None, align = 'edge', 
         color=['r', 'g', 'b'])
plt.show()
```

```python
labels = list(wine_type.index)
sizes = list(wine_type.values)
explode = (0.1, 0.05, 0.05)

# labels = A B C
# sizes = 파이 크기
# explode = 간격
# autopct = "%1.1f%%"

plt.title('"와인 등급 분포도"')
plt.pie(sizes, explode=explode, labels=labels, 
       autopct="%1.1f%%", shadow=True, startangle=250,
       colors = ['r', 'orange', 'yellow'])
plt.axis('equal')

plt.show()          
```

```python
plt.title('"와인 알코호올"')
# wine 에 있는 'alcohol 컬럼 이용'
# bins = 막대 몇개로 표현할건지 
# range = x 축 길이 
plt.hist('alcohol', bins = 13, range = (11,15), color='g',
        data = wine)
plt.show()
```
